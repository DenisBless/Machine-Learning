```python
# Imports
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
from scipy.stats import distributions as pdf
import matplotlib.pyplot as plt
import sklearn.datasets as dset
plt.style.use('seaborn')
```

```python
# Inits
start = np.eye(2)
M = np.array([[1,0.5],[0.4,0.8]])
U,S,V = np.linalg.svd(M)
S = np.eye(S.shape[0])*S
```

```python
# Plots I
f = plt.figure()
f.add_subplot(121)
f.set_figheight(5)
f.set_figwidth(10)
plt.xticks([])
plt.yticks([])
plt.title('Identity')
plt.quiver(0,0,*start, angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
f.add_subplot(122)
plt.xticks([])
plt.yticks([])
plt.title('M')
plt.quiver(0,0,*M, angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
plt.show()
```

```python
# Plots II
f = plt.figure()
f.add_subplot(141)
f.set_figheight(5)
f.set_figwidth(15)
plt.xticks([])
plt.yticks([])
plt.title('Identity')
plt.quiver(0,0,*start, angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)

f.add_subplot(142)
plt.xticks([])
plt.yticks([])
plt.title('V')
plt.quiver(0,0,*V, angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)

f.add_subplot(143)
plt.xticks([])
plt.yticks([])
plt.title(' SV')
plt.quiver(0,0,*(S@V), angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)

f.add_subplot(144)
plt.xticks([])
plt.yticks([])
plt.title('USV=M')
plt.quiver(0,0,*(U@S@V), angles='xy', scale_units='xy', scale=1)
plt.xlim(-1.2,1.2)
plt.ylim(-1.2,1.2)
plt.show()
```
