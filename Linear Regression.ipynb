```python
# Importing all libraries needed
import sklearn.datasets as dset
import torch
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

# Linear Regression

```python
def sin(num_samples, noise=0.3):
    x = np.linspace(-5,5,100)
    y = np.sin(x)
    samples = np.linspace (-5,5, num_samples)
    target =  np.random.choice((1-noise,1+noise), size=samples.size)*np.sin(samples)
    return x,y,samples,target

x,y,samples,target = sin(5)

plt.plot(x,y,c='b')
plt.scatter(samples,target,c='r')
plt.show()
```

When trying to fit a straight line to the data the Sum-of-Squares-Errorfunction is a good choice, since it minimizes the accumulated distance between the line and the data. The SSE-Loss is given by: <br>

$J(\mathbf{w}) = \frac{1}{2} \sum_{n=1}^N ( \sum_{j=0}^M x_{nj} w_j- y_n)^2 = \frac{1}{2} \sum_{n=1}^N (\mathbf{{x_{n}^T w}} - y_n)^2$

To find the parameter which are fitting the data best, the derivative of the loss with respect to the weights has to be computed. Because the MSE is a convex function, the global minima and therefore the best fit is guaranteed to be found. <br> 

$\frac{dJ(\mathbf{w})}{d\mathbf{w}}=
\begin{bmatrix}
\frac{\partial{J(w)}}{\partial{w_0}}\\
\vdots\\
\frac{\partial{J(w)}}{\partial{w_M}}\\
\end{bmatrix}
= 
\begin{bmatrix}
\sum_{n=1}^N ( \sum_{j=0}^M x_{nj} w_j- y_n)x_0\\
\vdots\\
\sum_{n=1}^N ( \sum_{j=0}^M x_{nj} w_j- y_n)x_M
\end{bmatrix} 
=
\sum_{n=1}^N ( \sum_{j=0}^M x_{nj} w_j- y_n)\mathbf{x}$

The amout of samples ('N' in the equation) can be adusted in the parameters. For ease regarding visualization, the demonstration uses only one feature ('M' in the equation)

```python
# Setting up the parameter of the model
param_dict = {
    'num_data_samples': 100,
    'noise': 60,
    'learning_rate': 1e-4,
    'num_epochs ': 400,
}
```

```python
x,y = dset.make_regression(n_samples=param_dict['num_data_samples'], n_features=1, noise=param_dict['noise'])
y /= 50  # Scaling y 
```

```python
# Adding bias term
bias = np.ones_like(x)
x = np.concatenate((bias,x), axis=1)
num_features = x.shape[1]
```

```python
# Visualizing the dataset
plt.scatter(x[:,1],y)
plt.xlabel('x: Data')
plt.ylabel('y: label/target')
plt.title("Dataset")
plt.show()
```

```python
# Initializing weights
w = 0.1*np.ones((num_features,1))
```

```python
# Mean Squared Error Loss
def SSE_loss(x,w,y):
    y_pred = np.dot(x,w).reshape(-1,) # Linear transformation
    loss = 0.5 * np.sum(np.square(y_pred-y), axis=0) 
    return loss

# Stochstic Gradient Descent
def optimzer(x,w,y):
    y_pred = np.dot(x,w).reshape(-1,) 
    dW = np.sum((y_pred - y)[:,None]*x , axis=0)
    w -= learning_rate*dW[:,None]
    return w
```

```python
learning_rate = param_dict['learning_rate']
loss_list = []
# Training the model:
for i in range(param_dict['num_epochs ']):
    loss = MSE_loss(x,w,y)
    loss_list.append(loss)
    w = optimzer(x,w,y)
```

```python
y_pred = np.dot(x,w)
plt.scatter(x[:,1],y)
plt.plot(x[:,1],y_pred, c='r')
plt.xlabel('x: Data')
plt.ylabel('y: label/target')
plt.title("Dataset")

plt.show()
```

```python
plt.plot(range(len(loss_list)), loss_list)
plt.xlabel('Number Epochs')
plt.ylabel('Sum-of-Squares Error')
plt.show()
```
